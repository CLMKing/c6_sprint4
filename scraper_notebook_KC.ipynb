```python
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
```

```python
# change the path to where your chromedriver is
PATH  = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
```

```python
# first 3 entries of IMDb's top 250
count = np.arange(1,4)
xpath_list = []
story_list = []

for i in count:
    xpath_list.append('//*[@id="main"]/div/span/div/div/div[3]/table/tbody/tr' + f"[{i}]" + '/td[2]/a')

# xpath list iterations
print(xpath_list)
```

```python
driver.get('https://www.imdb.com/chart/top?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=4da9d9a5-d299-43f2-9c53-f0efa18182cd&pf_rd_r=4Z3SA7T1WG5VJVTF6B3Z&pf_rd_s=right-4&pf_rd_t=15506&pf_rd_i=moviemeter&ref_=chtmvm_ql_3')

for j in xpath_list:
    search = driver.find_element_by_xpath(j)
    time.sleep(3)
    search.click()
    
    search_2 = driver.find_element_by_xpath('//*[@id="titleStoryLine"]/span[2]/a[1]')
    time.sleep(2)
    search_2.click()

    ps = driver.page_source
    soup = BeautifulSoup(ps, 'lxml')
    li_tag = soup.find_all('li', {'class':'ipl-zebra-list__item'})
    
    title = soup.find()

    story_list.append(li_tag[-1].get_text())
    print('Success! - '+ driver.title)
    time.sleep(3)
    driver.back()
    time.sleep(2)
    driver.back()

driver.close()
```

```python
# checking the length of the list (number of movies)
print(len(story_list))
```

```python
story_list = [x.replace('\n', '') for x in story_list]
story_list = [x.replace('\t', '') for x in story_list]
```
