```python
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np 
import requests
```

# Player Stats (Long Method)

```python
base_url = "https://www.pba.ph/stats"
```

```python
reqs = requests.get(base_url, headers = {'User-Agent':'Mozilla/5.0 (X11; Linux)'})
```

```python
soup = BeautifulSoup(reqs.content, 'lxml')
```

```python
stats = []
```

```python
table_tag = soup.find('table')
```

```python
tr_tag = table_tag.find_all('tr')
```

```python
for td in tr_tag: 
    td_tag = td.find_all('td')
    for i in td_tag:
        stats.append(i.get_text())
```

```python
player_name = stats[::23]
```
